```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

Question:- Build a recommender system by using cosine similarities score.

```python
data = pd.read_csv("book_1.csv",encoding = "latin1")
```

```python
data
```

```python
data.sort_values(by="User.ID")
```

```python
books = data.pivot_table(index='User.ID',
                            columns='Book.Title',
                            values='Book.Rating').reset_index(drop=True)
books
```

```python
books.index = data['User.ID'].unique()
```

```python
books
```

```python
books.fillna(0, inplace = True)
```

```python
books
```

```python
user = 1 - pairwise_distances( books.values,metric='cosine')
```

```python
user
```

```python
user_df = pd.DataFrame(user)
```

```python
user_df
```

```python
user_df.index = data['User.ID'].unique()
user_df.columns = data['User.ID'].unique()
```
